```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/tweets.csv')
```

```python
data.head()
```

```python
pd.set_option('display.max_colwidth', None)
messages = data[['label','tweet']]
messages.head()
```

```python
import gensim
```

```python
messages['tweet_clean_gensim'] = messages['tweet'].apply(lambda x : gensim.utils.simple_preprocess(x))
```

```python
messages.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(messages['tweet_clean_gensim'],messages['tweet'],test_size = 0.2, random_state = 42)
```

```python
w2v_model = gensim.models.Word2Vec(x_train,min_count=1,sg=1)
```

```python
w2v_model.wv['test']
```

```python
w2v_model.wv.most_similar('test')
```

```python
words = w2v_model.wv.index_to_key
```

```python
len(words)
```

```python
x_train_vec = np.array([np.array([w2v_model.wv[i] for i in ls if i in words]) for ls in x_train],dtype=object)
x_test_vec = np.array([np.array([w2v_model.wv[i] for i in ls if i in words]) for ls in x_test],dtype=object)
```

```python
x_train_vec_avg = []
for v in x_train_vec:
  if v.size:
    x_train_vec_avg.append(v.mean(axis = 0))
else:
  x_train_vec_avg.append(np.zeros(100,dtype=float))
x_test_vec_avg = []
for v in x_test_vec:
  if v.size:
    x_test_vec_avg.append(v.mean(axis = 0))
  else:
    x_test_vec_avg.append(np.zeros(100,dtype=float))
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)
clf.fit(x_train_vec_avg,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(x_train_vec_avg)
accuracy_score(y_pred,y_train)
```

```python
y_pred = clf.predict(x_test_vec_avg)
accuracy_score(y_pred,y_test)
```
